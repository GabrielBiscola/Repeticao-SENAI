```python
# Um funcionário de uma empresa recebe aumento salarial anualmente:
#   Sabe-se que:
#     Esse funcionário foi contratado em 1995, com salário inicial de R$ 1.000,00;
#     Em 1996 recebeu aumento de 1,5% sobre seu salário inicial;
#     A partir de 1997 (inclusive), os aumentos salariais sempre correspondem ao dobro do percentual do ano anterior.
# Faça um programa que determine o salário atual desse funcionário.
# Após concluir isto, altere o programa permitindo que o usuário digite o salário inicial do funcionário.
si = float(input("Salário inicial:"))
for i in range (2024-1995):
  rise = (1.5/100)*(2**i)
  si *= rise+1
  print(f"Salário em {1995+(i+1)}: R${si:.2f}") #aumento é uma palavra mto grande e raise é alguma função 0,15 0,3 0,6 1,2
```

```python
# Uma loja tem tem uma política de descontos de acordo com o valor da compra do cliente.
# Os descontos começam acima dos R$500. A cada 100 reais acima dos R$500,00 o cliente ganha 1% de desconto cumulativo até 25%.
# Por exemplo: R$500 = 1% || R$600,00 = 2% … etc…
# Faça um programa que exiba essa tabela de descontos no seguinte formato:
# Valordacompra – porcentagem de desconto – valor final
for i in range (500, 500+(100*250), 100):
  desconto = (i-400)//100
  porcent = desconto/100
  if desconto <= 25:
    print(f"{i} - {desconto}% - {i*(1-porcent):.2f}")
```

```python
# Faça um programa que peça dois números, base e expoente, calcule e mostre o primeiro número elevado ao segundo número. Não utilize a função de potência da linguagem.
base = int(input())
expoente = int(input())
result = base
if expoente == 0:
  result = 1
else:
  for i in range (expoente-1):
    result *= base
print(f'\n {expoente}\n{base}\t=\t{result}')
```

```python
# Faça um programa que mostre todos os primos entre 1 e N sendo N um
# número inteiro fornecido pelo usuário. O programa deverá mostrar também
# o número de divisões que ele executou para encontrar os números primos.
# Serão avaliados o funcionamento, o estilo e o número de testes (divisões)
# executados.
stop = False
par = int()
primos = []
nlist=[]
count = 0
nprimo = []
n = int(input())
if n > 2:
  if n != 3:
    for i in range (1, n+1):
      nlist.append(i)
    for i in range (2, n, 2):
      primos.append(nlist[i]) #tenho todos os ímpares
      if i+2 > n:
        break
    for i in primos:
      for j in range (2, i//2):
        count += 1
        if i%j == 0:
          nprimo.append(i)
          break
    for i in nprimo:
      primos.remove(i)
  else:
    primos = [2]

else:
  print("0 números primos.")
primos.insert(0, 2)
print(count, "divisões")
print(primos)
```
