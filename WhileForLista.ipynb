Solicitar um numero inteiro positivo N ao usuario. Imprimir os numeros de N a 1. Realizar o exercicio usando loops for e while


```python
T = int(input("Entre com numero inteiro maior que 0: "))
#FOR
print("For:")
for t in range (T):
  print(f"{T-t}", end=" ")
print()

#While
print("While:")
i = 0
while i < T:
  print(f"{T-i}", end=" ")
  i+=1
```

Solicitar um número inteiro positivo N ao usuário. Imprimir os números de N a 1 trocando o sinal dos números. Realizar o exercício usando loops for e while. Exemplo para N igual a 5:

```python
T = int(input("Entre com numero inteiro maior que 1: "))
#FOR
print("For:")
mult = -1
sinal = 1
for t in range (T):
  print(f"{(T-t)*sinal}", end=" ")
  sinal *= mult
print()

#While
sinal = 1
print("While:")
i = 0
while i < T:
  result = sinal*(T-i)
  print(f"{result}", end=" ")
  i+=1
  sinal *= mult
```

```python
pali = bool()
palavra = str(input("Entre ai com alguma palavra")).strip()
if len(palavra) > 2:
  for i in range (1, len(palavra)+1):
    if palavra[i-1] != palavra[-i]:
      pali = False
      break
    else:
      pali = True
if pali:
  print(f"{palavra} é um palíndromo")
else:
  print(f"{palavra} não é um palíndromo")
```

```python
cont = 0
stop = False
n = int(input("Crie um padrão:"))
while stop != True:
  print(" " + n * "-" + " ")
  if cont == n:
    stop = True
    break
  for j in range (n):
    print("|" + n * " " + "|")
    cont += 1
```

```python
#For
n = int(input("Entre com algum número:"))
for i in range (n+1):
  print("+" * i)
  print()

#While
i = 0
while i != n+1:
  print("+" * i)
  print()
  i += 1
```
