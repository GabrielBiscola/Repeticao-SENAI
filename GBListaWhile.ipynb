```python
precos = []
while True:
  item = 1
  precos.clear()
  while True:
    precos.append(float(input(f"Preco do produto {item}:")))
    if precos[-1] == 0:
      precos.pop()
      break
    item += 1
  total = sum(precos)
  pago = float(input("Quanto voce pagou?"))
  troco = pago-total
  if troco < 0:
    print("CALOTEIRO!")

    break
  print("\n\t Lojas Tabajara")
  for i in range (1, len(precos)+1):
    print(f"Produto {i}: {precos[i-1]}")
  print(f"Total: R${total:.2f}\nDinheiro: R${pago:.2f}\nTroco: R${(troco):.2f}\n")
```

```python
#11
option = ""
fim = "n"
while option != "0": #Tela inicial
  options = [0,1,2]
  nome = ["Cachorro Quente", "Bauru Simples", "Bauru com Ovo", "Hamburguer", "Cheeseburguer", "Refrigerante"]
  cod = [100,101,102,103,104,105]
  preco = [1.2,1.3,1.5,1.2,1.3,1]
  encerrar = False
  quantidade = [0,0,0,0,0,0]
  option = ""
  total = 0

  print("0 - Sair\n1- Menu\n2- Fazer pedido")
  while option not in options:
    option = int(input()) #selecao opcao

  if option == 2: #Comprar
    while encerrar != True:
      pedido=int(input("Digite o codigo do item de sua escolha:"))%100
      quantidade[pedido]=(int(input("Quantos voce quer?")))
      fim = str(input("Algo mais? (s/n)")).lower()
      if fim == "n":
        print("\n\tSeu pedido:")
        for i in range(0, len(quantidade)):
          if quantidade[i] == 0:
            continue
          else:
            print(f"{quantidade[i]} {nome[i]} - R${preco[i]*quantidade[i]:.2f}")
            total += float(preco[i]*quantidade[i])
        print(f"Total = R${total:.2f}\n")
        encerrar = True
      #finalizar pedido

  elif option == 1: #Menu
    print("Nome\t\tCódigo\tPreço")
    for i in range (0, len(nome)):
      print(f"{nome[i]}\t{cod[i]}\t{preco[i]:.2f}")
    if (i == len(nome)-1):
      print("")

  else: #encerrar programa
    break
```
